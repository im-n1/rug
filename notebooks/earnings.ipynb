```python
import httpx
from datetime import date, timedelta
import time
```

```python
from_date = date.today()
to_date = date.today() + timedelta(days=7)

# Check date span.
if 14 < (to_date - from_date).days:
    raise ValueError("Maximum date span is 14 days.")
    
response = httpx.get(f"https://api.stocktwits.com/api/2/discover/earnings_calendar?date_from={from_date}&date_to={to_date}")
earnings = response.json()["earnings"]
data = []

for day, item in earnings.items():
    for stock in item["stocks"]:
        data.append({
            "date": date(*time.strptime(stock["date"], "%Y-%m-%d")[:3]),
            "symbol": stock["symbol"],
            "time": stock["time"],
            "when": "before_open" if "08:00:00" == stock["time"] else "after_close",
        })
    
data
```
